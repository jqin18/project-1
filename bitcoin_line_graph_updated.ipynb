```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import sys
import time
import datetime

from statistics import mode

import matplotlib.dates as mdates


bitcoin_path = "/Users/minoperic/Data Analytics/Project 1/Data/bitcoin_cleaned.csv"

bitcoin_df = pd.read_csv(bitcoin_path,low_memory=False)

bitcoin_df['trade_date'] = pd.to_datetime(bitcoin_df.trade_date)

bitcoin_df.head(10)
```

```python
# Generate line plot of date (x) vs. price (y)

# plot data
fig, ax = plt.subplots(figsize=(15,7))

x=bitcoin_df["trade_date"]
y=bitcoin_df["price_usd"]


# use unstack()
bitcoin= plt.plot(x,y, color="blue" )

plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.title("Yearly change of Bitcoin (Jan 2016 - Aug 2020)")

plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
```
