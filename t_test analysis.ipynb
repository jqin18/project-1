```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import scipy.stats as stats

crypto_path = "resources/cleaned_crypto.csv"

crypto_data = pd.read_csv(crypto_path,low_memory=False)
```

```python
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
#independent t-test (crypto_type)

coin = crypto_data[crypto_data["crypto_type"] == 0]
coin.market_cap.mean()
```

```python
token = crypto_data[crypto_data["crypto_type"] == 1]
token.market_cap.mean()
```

```python
stats.ttest_ind(coin.market_cap, token.market_cap, equal_var=False)
```

```python
#boxplot

crypto_data.boxplot("market_cap", by="crypto_type", figsize=(20, 10))
```

```python
#independent t-test (sourcing_type)
mineable = crypto_data[crypto_data["minable"] == 1]
mineable.market_cap.mean()
```

```python
not_mineable = crypto_data[crypto_data["minable"] == 0]
not_mineable.market_cap.mean()
```

```python
stats.ttest_ind(mineable.market_cap, not_mineable.market_cap, equal_var=False)
```

```python
#boxplot

crypto_data.boxplot("market_cap", by="minable", figsize=(20, 10))
```
