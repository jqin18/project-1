```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import sys
import time
import datetime

from statistics import mode

import matplotlib.dates as mdates


bonds_path = "/Users/minoperic/Data Analytics/Project 1/Data/government_bonds.csv"

bonds_df = pd.read_csv(bonds_path,low_memory=False)

bonds_df['trade_date'] = pd.to_datetime(bonds_df.trade_date)

bonds_df.head(10)
```

```python
# Generate line plot of date (x) vs. price (y)

# plot data
fig, ax = plt.subplots(figsize=(15,7))

x=bonds_df["trade_date"]
y=bonds_df["bond_rate"]


# use unstack()
bitcoin= plt.plot(x,y, color="red" )

plt.xlabel("Date")
plt.ylabel("Interest Rate (2 Year Yield)")
plt.title("Yearly change of RBA Government Bond Rate - 2 Year Yield (Jan 2016 - Aug 2020)")

plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
```
